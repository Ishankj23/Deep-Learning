```python
import pandas as pd
url = "https://raw.githubusercontent.com/utkarshg1/Neural-Net-Datasets/main/booking_train.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
x=df.drop(columns=['id', 'booking_status'])
y=df[['booking_status']]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
con_pipe=Pipeline(steps=[('impute', SimpleImputer(strategy='median')),
                         ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
x_pre=con_pipe.fit_transform(x)
x_pre.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest= train_test_split(x_pre, y, test_size=0.2, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Input, Dropout
```

```python
model=Sequential()
model.add(Input(shape=(xtrain.shape[1],)))
model.add(Dense(units=64, activation='relu'))
# dropout layer 1
model.add(Dropout(rate=0.4))
model.add(Dense(units=32, activation='relu'))
model.add(Dropout(rate=0.3))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
hist = model.fit(xtrain, ytrain, validation_split=0.2, epochs=60)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.plot(hist.history['loss'], label='Train Loss')
plt.plot(hist.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Binary Crossentropy')
plt.title('Learning Curve')
plt.legend()
plt.grid()
plt.show()
```

```python
model.evaluate(xtrain, ytrain)
```

```python
model.evaluate(xtest, ytest)
```

```python
yprob_train = model.predict(xtrain)
yprob_test = model.predict(xtest)
```

```python
yprob_train[0:5]
```

```python
yprob_test[0:5]
```

```python
ypred_train = [1 if prob>=0.5 else 0 for prob in yprob_train]
ypred_test = [1 if prob>=0.5 else 0 for prob in yprob_test]
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cf = confusion_matrix(ytest, ypred_test)
cfd = ConfusionMatrixDisplay(cf)
cfd.plot()
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_test))
```
